# 4. Modelling

```python
sns.countplot(x='quality',data=df)
```

```python
# converting the response variables(3-7) as binary response variables that is either good or bad

df['quality'] = df['quality'].map({3 : 'bad', 4 :'bad', 5: 'bad',
                                      6: 'good', 7: 'good', 8: 'good', 9: 'good'})
```

```python
df.tail()
```

```python
# analyzing the different values present in the dependent variable(quality column)
df['quality'].value_counts()
```

```python
from sklearn.preprocessing import LabelEncoder

le = LabelEncoder()

df['quality'] = le.fit_transform(df['quality'])
```

```python
df.tail()
```

```python
df['quality'].nunique()
```

```python
df['quality'].unique()
```

```python
sns.countplot(df['quality'])
```

```python
# dividing the dataset into dependent and independent variables

x = df.iloc[:,:11]  # predictor variable
y = df.iloc[:,11]   # target variable 

# determining the shape of x and y.
print(x.shape)
print(y.shape)
```

```python
# dividing the dataset in training and testing set

from sklearn.model_selection import train_test_split

x_train, x_test, y_train, y_test = train_test_split(x, y, test_size = 0.25, random_state = 42)

# determining the shapes of training and testing sets
print(x_train.shape)
print(y_train.shape)
print(x_test.shape)
print(y_test.shape)
```

```python
# standard scaling 
from sklearn.preprocessing import StandardScaler

sc = StandardScaler()
x_train = sc.fit_transform(x_train)
x_test = sc.fit_transform(x_test)
```

## use sklearn Logistic Regression

```python
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import classification_report, confusion_matrix
from sklearn.model_selection import GridSearchCV, cross_val_score



# creating the model
model = LogisticRegression()

# feeding the training set into the model
model.fit(x_train, y_train)

# predicting the results for the test set
y_pred = model.predict(x_test)

# calculating the training and testing accuracies
print("Training accuracy :", model.score(x_train, y_train))
print("Testing accuracy :", model.score(x_test, y_test))

```

## Evaluate your model 

```python
# classification report
print(classification_report(y_test, y_pred))
```

```python
# confusion matrix
cf_matrix = confusion_matrix(y_test, y_pred)
print(cf_matrix)
```

```python
ax = sns.heatmap(cf_matrix, annot=True, cmap='Blues', fmt="d")
ax.set_title('Seaborn Confusion Matrix with labels\n\n');
ax.set_xlabel('\nPredicted quality')
ax.set_ylabel('Actual quality ');

## Ticket labels - List must be in alphabetical order
ax.xaxis.set_ticklabels(['bad','good'])
ax.yaxis.set_ticklabels(['bad','good'])

## Display the visualization of the Confusion Matrix.
plt.show()
```
